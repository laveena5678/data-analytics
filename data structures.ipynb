```python
sales_data = [
    ("Electronics", 50000),
    ("Fashion", 30000),
    ("Electronics", 70000),
    ("Home Decor", 20000),
    ("Fashion", 50000),
]

new_entry = ("Home Decor", 15000)

sales_data.append(new_entry)

update_data = {}

for products,sales in sales_data:
    if products in update_data:    
        update_data[products] += sales
    else:
        update_data[products] = sales

print(update_data)
```

```python
# Defining a database schema for a customer dataset
database_schema = {
    "Customer_ID": "int",
    "Name": "Str",
    "Age": "Int",
    "City": "Str",
    "Spending": "Float"
}

# Sample customer data record
customer_record = {"Customer_ID": 101, "Name": "Ravi", "Age": 28, "City": "Delhi", "Spending": 75000.0}

# Validating data against schema
for column, expected_type in database_schema.items():
    actual_type = type(customer_record[column]).__name__
    if actual_type.lower() != expected_type.lower():
        print(f"Error: {column} expected {expected_type}, but got {actual_type}.")
    else:
        print(f"{column} is valid.")
```
