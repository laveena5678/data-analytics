 Phone Number Validation in Python

```python
import re

def is_valid_phone(phone):
    pattern = r"^[6-9]\d{9}$" 
    return bool(re.match(pattern, phone))


phone_numbers = ["9876543210", "12345", "9988776655", "abcd123456", "7778889990"]

for phone in phone_numbers:
    if not is_valid_phone(phone):
        print(f"Invalid phone number detected: {phone}")
```

Detecting Invalid Salary Values

```python
import pandas as pd

# Simulated dataset
df = pd.DataFrame({
    "Employee": ["Raj", "Meera", "Akash", None],
    "Salary": ["50000", "-20000", "FiftyK", "70000"]
})

# Detect null values
if df.isnull().values.any():
    print("Null values found in dataset.")

# Validate salary values

for index, row in df.iterrows():
    try:
        salary = float(row["Salary"])  
        if salary < 0:
            print(f"Invalid salary for {row['Employee']}: Negative salary not allowed.")
    except ValueError:
        print(f"Invalid salary format for {row['Employee']}: {row['Salary']}")
```

validated email and age

```python
import re

def validate_user(email, age):
    email_pattern = r"^[\w\.-]+@[\w\.-]+\.\w+$"
    
    if not re.match(email_pattern, email):
        print(f"Invalid email: {email}")
    
    try:
        age = int(age)
        if age <= 0 or age > 120:  # Assuming realistic human age limits
            print(f"Invalid age: {age}")
    except ValueError:
        print(f"Invalid age format: {age}")


validate_user("john.doe@example.com", "25")  
validate_user("invalid-email", "30")  
validate_user("jane@domain", "twenty") 
validate_user("peter.parker@mail.com", "-5")  
validate_user("bruce.wayne@wayne.com", "150")
```

Cleaning Salary Data with Try-Except

```python
def clean_salary(salary):
    try:
        salary = float(salary)  
        if salary < 0:
            raise ValueError("Salary cannot be negative.")  
        return salary
    except ValueError as e:
        print(f"Invalid salary format: {salary}. Replacing with 0. Error: {e}")
        return 0

salaries = ["50000", "-20000", "FiftyK", "70000"]
cleaned_salaries = [clean_salary(salary) for salary in salaries]

print(cleaned_salaries)
```

Example: Handling and Logging JSON Parsing Errors

```python
import json
import logging

logging.basicConfig(filename="json_processing.log", level=logging.ERROR, format="%(asctime)s - %(levelname)s - %(message)s")

def load_json(data):
    try:
        return json.loads(data)  
    except json.JSONDecodeError as e:
        logging.error(f"Invalid JSON data: {data} | Error: {e}")
        return {}  

# Sample JSON inputs
json_data_list = [
    '{"name": "Alice", "age": 25}',  
    '{"name": "Bob", "age": "twenty-five"}',  
    '{name: "Charlie", age: 30}',  
    '{"name": "David", "age": }'  
]

# Process each JSON entry
parsed_jsons = [load_json(data) for data in json_data_list]

print(parsed_jsons)  # Output parsed JSONs
```


Example: Handling Missing Age Data in a Customer Database

```python
import pandas as pd
import numpy as np


data = {
    "Customer": ["Alice", "Bob", "Charlie", "David"],
    "Age": [28, None, 35, None]  
}

df = pd.DataFrame(data)


try:
    avg_age = np.nanmean(df["Age"])  
    df["Age"].fillna(avg_age, inplace=True)  
except Exception as e:
    print(f"Error while imputing missing age: {e}")

print(df)
```

Example: Custom Exception Handling in an E-Commerce System

```python
class MissingProductIDError(Exception):
    """Raised when a product ID is missing in an order."""
    pass

class InvalidQuantityError(Exception):
    """Raised when an order quantity is zero or negative."""
    pass

def validate_order(order):
    if order.get("ProductID") is None:
        raise MissingProductIDError("ProductID is missing in order record.")
    
    if order.get("Quantity", 0) <= 0:
        raise InvalidQuantityError("Quantity must be greater than zero.")

# Simulated order processing
orders = [
    {"ProductID": None, "Quantity": 2},  # Missing ProductID
    {"ProductID": "P12345", "Quantity": 0},  # Invalid Quantity
    {"ProductID": "P67890", "Quantity": 5}  # Valid Order
]

for order in orders:
    try:
        validate_order(order)
        print(f"Order {order} is valid.")
    except (MissingProductIDError, InvalidQuantityError) as e:
        print(f"Validation error caught: {e}")
```

Z-Score विधि

```python
import numpy as np


delivery_times = [3, 4, 2.5, 3.2, 5, 6, 30, 0.5, 4.5, 3.8, 7]

mean = np.mean(delivery_times)
std = np.std(delivery_times)


for time in delivery_times:
    z_score = (time - mean) / std
    if abs(z_score) > 2: 
        print(f"Outlier detected (Z-score method): {time}")
```

Interquartile Range (IQR) विधि

```python
import pandas as pd


df = pd.Series(delivery_times)


Q1 = df.quantile(0.25)  
Q3 = df.quantile(0.75) 
IQR = Q3 - Q1  


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


outliers = df[(df < lower_bound) | (df > upper_bound)]
for outlier in outliers:
    print(f"Outlier detected (IQR method): {outlier}")
```


उदाहरण: डेटा ट्रांसफॉर्मेशन से फ़ॉर्मेटिंग एरर ठीक करना

```python
import pandas as pd


df = pd.DataFrame({"Quantity Sold": ["1,200 units", "500 pcs", "2.5K", "1000"]})


def clean_quantity(value):
    value = value.lower().replace(",", "")  
    value = value.replace(" units", "").replace(" pcs", "")  
    
    if "k" in value:  
        value = value.replace("k", "")  
        return float(value) * 1000  
    return float(value)  

df["Quantity Sold"] = df["Quantity Sold"].apply(clean_quantity).astype(int)  
print(df)
```

```python
from datetime import datetime


def validate_student(student):
    if student["Age"] < 18 or student["Age"] > 60:
        raise ValueError(f"Invalid age ({student['Age']}). Age must be between 18 and 60.")
    
    if student["GPA"] < 0.0 or student["GPA"] > 4.0:
        raise ValueError(f"Invalid GPA ({student['GPA']}). GPA must be between 0.0 and 4.0.")
    
    if student["Admission Date"] > datetime.today().date():
        raise ValueError(f"Invalid admission date ({student['Admission Date']}). Date cannot be in the future.")


student = {
    "Name": "Rahul",
    "Age": 17,  
    "GPA": 3.8,
    "Admission Date": datetime(2023, 5, 10).date()
}


try:
    validate_student(student)
except ValueError as e:
    print(f"Validation error: {e}")
```
