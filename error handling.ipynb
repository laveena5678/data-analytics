Handling Missing Data: Filling with Default Values


```python
import pandas as pd

data = {"Customer": ["Ravi", "Aisha", "Priya", "Deepak"],
        "Revenue": [75000, None, "Thirty Thousand", 15000]}

df = pd.DataFrame(data)

# Function to fill missing data
def handle_missing_data(value):
    try:
        return int(value)
    except (ValueError, TypeError):
        return 0  # Replace with 0 if data is missing or invalid

df["Revenue"] = df["Revenue"].apply(handle_missing_data)
print(df)
```

 Handling Missing Values (None) in Dataset


```python
import pandas as pd

data = {"Name": ["Ravi", "Aisha", None, "Deepak"],
        "Age": [25, None, 30, 40]}

df = pd.DataFrame(data)

# Handling missing values
df.fillna("Unknown")  # Replacing None with "Unknown"
print(df)
```

 Converting String Data to Numeric Values

```python
data = {"Revenue": ["1000", "2000", "three thousand", "4000"]}

df = pd.DataFrame(data)

# Converting strings to integers, handling errors
def convert_to_numeric(value):
    try:
        return int(value)
    except ValueError:
        print(f"Skipping invalid value: {value}")
        return 0  # Default value for invalid data

df["Cleaned_Revenue"] = df["Revenue"].apply(convert_to_numeric)
print(df)
```

 Handling Non-Numeric Data in Calculations

```python
import pandas as pd

data = {"Product": ["Laptop", "Phone", "Tablet"],
        "Sales": ["50000", "Twenty Thousand", "30000"]}

df = pd.DataFrame(data)

def clean_sales(value):
    try:
        return int(value)
    except ValueError:
        print(f"Skipping invalid value: {value}")
        return 0  # Replacing incorrect values with 0

df["Cleaned_Sales"] = df["Sales"].apply(clean_sales)
print(df)
```

Handling Division by Zero Error

```python
def calculate_profit(sales, expenses):
    try:
        profit = sales / expenses
    except ZeroDivisionError:
        print("Error: Division by zero")
        return 0  
    return profit

print(calculate_profit(1000, 0))  
```

Safe Division to Prevent ZeroDivisionError

```python
def safe_division(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Division by zero"

print(safe_division(100, 0))
print(safe_division(100, 5))
```

Safe File Handling

```python
file_path="customer_reviews.txt"
def read_file(file_path):
    try:
        with open(file_path, "r") as file:
            data = file.read()
        return data
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return None  

print(read_file("customer_reviews.txt"))
```

Handling Multiple Errors in a Single Try-Except

```python
def process_data(value):
    try:
        result = 100 / value
        print(f"Result: {result}")
    except (ZeroDivisionError, TypeError) as e:
        print(f"Error: {e}")

process_data(10)  # Test with zero
process_data("abc")  # Test with invalid type
```

 Handling Incorrect Data Types in Dataset

```python
import pandas as pd

data = {"ID": ["101", "102", "ABC", "104"]}

df = pd.DataFrame(data)

def convert_to_int(value):
    try:
        return int(value)
    except ValueError:
        return None

df["Converted_ID"] = df["ID"].apply(convert_to_int)
print(df)
```

Handling ValueError in Data Type Conversion

```python
def convert_age(age_str):
    try:
        return int(age_str)
    except ValueError:
        print(f"Error: '{age_str}' is not a valid number.")
        return None  # Return None for invalid data

ages = ["25", "30", "invalid_age", "35"]
converted_ages = [convert_age(age) for age in ages]
print(converted_ages)
```

 Handling KeyError in DataFrame Operations

```python
import pandas as pd

data = {"Name": ["Amit", "Neha"], "Age": [28, 22],"Salary":[34000,89000]}
df = pd.DataFrame(data)

try:
    print(df["Salary"])  # Column "Salary" doesn't exist
except KeyError:
    print("Error: Column 'Salary' not found in DataFrame")
```

Using else and finally with Try-Except

```python
def read_and_process_file(file_path):
    try:
        with open(file_path, "r") as file:
            data = file.read()
        print("File read successfully")
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
    else:
        print("Processing data after reading the file.")
    finally:
        print("End of file handling.")

read_and_process_file("customer_reviews.txt")
```

Handling API Errors in Data Fetching

```python
import requests

try:
    response = requests.get("https://randomuser.me/api/")
    response.raise_for_status()  # Raises an error for bad response
    print(response.json())
except requests.exceptions.RequestException as e:
    print(f"Error fetching data: {e}")
```

Handling Index Errors with List Operations

```python
def safe_access_element(lst, index):
    try:
        return lst[index]
    except IndexError:
        print(f"Error: Index {index} out of range.")
        return None  # Return None if index is invalid

my_list = [10, 20, 30,40,50,70]
print(safe_access_element(my_list, 5))  # Invalid index
```

Handling File I/O Errors During Writing

```python
def write_data_to_file(file_path, data):
    try:
        with open(file_path, "w") as file:
            file.write(data)
        print("Data written successfully")
    except IOError:
        print("Error: There was an issue writing to the file.")

write_data_to_file("customer_reviews.txt", "This is some test data.")
```

Handling ValueError in Data Parsing

```python
import pandas as pd

data = {"Date": ["2024-03-20", "InvalidDate", "2024-03-22"]}

df = pd.DataFrame(data)

def parse_date(value):
    try:
        return pd.to_datetime(value)
    except ValueError:
        return None  # Replace invalid dates with None

df["Parsed_Date"] = df["Date"].apply(parse_date)
print(df)
```
